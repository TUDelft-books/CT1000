:class: tip, dropdown

In $\text{D}$ the shear stress $0$.

Relevant cross-sectional properties are:
- $A = 20250 \text{ mm}^2$
- From top fibre downwards, the normal force centre is at $70 \text{ mm}$
- $I_{zz} = 51300000 \text{ mm}^4$

The normal stress is $-25 \text{ MPa}$ with no shear stresses.

```{figure} intro_data/xy.svg
:align: center
```

As this is the only stress, the principal stresses are: $\sigma_1 = -25 \text{ MPa}$, $\sigma_2 = \sigma_3 = 0 \text{ MPa}$.

Because of $\sigma_2 = 0 \text{ MPa}$, the maximum shear stress is $\cfrac{25}{2} = 12.5 \text{ MPa}$ at $45^{\text{o}}$. Or it can be found by rotating the right plane $\cfrac{1}{4}\pi \text{ rad}$:

- $\sigma_{\bar x \bar x} = -25 \cos \left(\cfrac{1}{4}\pi\right)^2 = -12.5 \text{ MPa} $
- $\tau_{\bar x \bar y} = -25 \sin \left(\cfrac{1}{4}\pi\right) \cos \left(\cfrac{1}{4}\pi\right) = 12.5 \text{ MPa}$

```{figure} intro_data/rotated.svg
:align: center
```

`````python
import sympy as sym
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']

h, q, N, L, b, h_A = sym.symbols('h q N L b h_A',real=True, positive=True)	

h = sym.nsimplify(180)
b = sym.nsimplify(150)
L = 6
q = sym.nsimplify(45.6)
h_A = sym.nsimplify(50)
N = 101.25

M =  q * L**2 / 8

tau = 0
A = b * h - b * h/2 /2
display(A)
zc = ( b * h/2 * h/4 + b * ( h/2 ) / 2 * h/3*2 ) / A
display(zc)
Izz = b * (h/2)**3 / 12 + b * (h/2) * (h/2 - zc)**2 + b * (h/2)**3 / 36 + b * (h/2) * (h/6*4 - zc)**2
display(Izz)
sigma = -M * 100000 * h_A / Izz - N * 1000 / A
display(sigma)

sigma_min = (sigma)/2-sym.sqrt(((-sigma)/2)**2+tau**2)
sigma_max = (sigma)/2+sym.sqrt(((-sigma)/2)**2+tau**2)
#display(sigma_min)
#display(sigma_max)
alpha_0 = sym.atan2(-tau,-sigma/2)/2
display((alpha_0*180/sym.pi).evalf())
#print(np.rad2deg(alpha_tau))

f_y = sym.sqrt(((sigma_max - sigma_min)**2 + sigma_max **2 + sigma_min**2)/2)
#display(f_y)
f_y_2 = sym.sqrt(sigma**2 + 3 * tau**2)
#display(f_y_2)
#display(f_y_2.evalf())

alpha = sym.pi  / 4
sigma_xx = sigma * sym.cos(alpha)**2
tau_xy = - sigma*sym.sin(alpha)*sym.cos(alpha)
display(sigma_xx)
display(tau_xy)

plt.figure()
ax = plt.gca() #you first need to get the axis handle
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_label_coords(1.05, 0.5)
ax.yaxis.set_label_coords(0,0.95)
#plt.axis([-2e8,12e8,-8e7, 8e7])
plt.ylim(-15,15)
plt.grid()
plt.gca().set_aspect("equal")
plt.plot(sigma,tau,marker='o')
plt.plot(0,-tau,marker='o')
plt.plot([sigma,0],[tau,-tau])
plt.plot((sigma)/2,0,marker='o')
circle1 = plt.Circle([(sigma)/2,0],sym.sqrt(((sigma)/2)**2+tau**2),fill=False)
ax.add_artist(circle1)
plt.plot(sigma_min,0,marker='o')
plt.plot(sigma_xx,-tau_xy,marker='o',color='black')
plt.plot([sigma_xx,0],[-tau_xy,-tau],color='black')
plt.gcf().patch.set_alpha(0)

# Save the figure with transparent background
#plt.savefig('plot.svg', format='svg', transparent=True);
`````
