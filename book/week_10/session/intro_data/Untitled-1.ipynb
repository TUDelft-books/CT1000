```python
import sympy as sym

L1, L2, EA = sym.symbols('L1 L2 EA')
L1 = sym.S(9)/sym.S(2)
L2 = sym.S(6)
EA = sym.nsimplify(22.5*2*1000)
h = L1 / 3 * 4

w_B = sym.Symbol('w_B')
w_B = sym.nsimplify(0.03)
B_v = sym.Symbol('B_v')
```

```python
N_BD = B_v
N_AD = N_BD / 4 * 5
N_CD = N_AD / 5 * 3

display(N_BD, N_AD, N_CD)

deltaL_CD = N_CD / EA * L2
deltaL_AD = N_AD / EA * L1 / 3 * 5
deltaL_BD = B_v / EA * L1 / 3 * 4
display(deltaL_CD, deltaL_AD, deltaL_BD)

x_D = deltaL_CD
z_D = deltaL_AD/5*4 + (deltaL_AD/5*3 + deltaL_CD) /4 * 3
z_B = z_D + deltaL_BD
display(z_B)
Eq1 = sym.Eq(z_B, w_B)

B_v_sol = sym.solve(Eq1, B_v)[0]
display(B_v_sol)
```

```python
display(N_BD.subs(B_v, B_v_sol))
display(N_AD.subs(B_v, B_v_sol))
display(N_CD.subs(B_v, B_v_sol))
```

```python
display(x_D.subs(B_v, B_v_sol).evalf())
display(z_D.subs(B_v, B_v_sol).evalf())
```

```python
N_BD_fout = w_B / L2 * EA
N_AD_fout = N_BD_fout / 4 * 5
N_CD_fout = N_AD_fout / 5 * 3

display(N_BD_fout, N_AD_fout.evalf(), N_CD_fout.evalf())

deltaL_AD = N_AD_fout / EA * L1 / 3 * 5
deltaL_CD = N_CD_fout / EA * L2

display(deltaL_AD.evalf(), deltaL_CD.evalf())

display((deltaL_AD/5*4).evalf())
```
