```python
import sympy as sym
```

```python
EI = sym.symbols('EI')
EA = sym.symbols('EA')

EI = sym.nsimplify(20e3)
EA = sym.nsimplify(10e3)
q = sym.symbols('q')
q = 8 #sym.S(5)
q_sol = 8 #5
N_CD = sym.symbols('N_CD')

deltaT_AC_eps = 50
deltaT_AC_kap = 8
deltaT_CD = 50
h = sym.nsimplify(0.4)
alpha = sym.nsimplify(1e-5)
```

```python
kappa_AC = sym.nsimplify(1e-5) * deltaT_AC_kap / h
display(kappa_AC, kappa_AC.evalf())

eps_AC = sym.nsimplify(1e-5) * deltaT_AC_eps

display(eps_AC, eps_AC.evalf())
```

```python
T, N = sym.symbols('T N')

T_sol = EI * kappa_AC
N_sol = EA * eps_AC
display(T_sol, T_sol.evalf())
display(N_sol, N_sol.evalf())
```

```python
M_B = -(q * 6 * 6 / 2 + N_CD * 6 + T)
M_BC = - ( q * 3 * 3 /2 + N_CD * 3 + T)
M_BC_hulp = M_BC + q * 6**2 / 8
display(M_B, M_B.evalf())
display(M_B.subs(T,T_sol), M_B.subs(T,T_sol).evalf())
phi_A = -M_B * 4 / 6 / EI + T * 4 / 3 / EI
phi_B = -M_B * 4 / 3 / EI + T * 4 / 6 / EI
w_AB = M_B * 4 **2 / EI / 16 - T * 4 **2 / 16 / EI
w_CD = T * 4 **2 / 16 / EI + T * 4 **2 / 16 / EI
display(phi_B, phi_B.evalf())
display(phi_B.subs(T,T_sol), phi_B.subs(T,T_sol).evalf())
w_C_BC = - (phi_B * 6 + T * 6 **2 / 2 / EI + N_CD * 6**3 / 3 / EI + q * 6**4 / 8 / EI)
display('w_C_BC')
display(w_C_BC, w_C_BC.evalf())
display(w_C_BC.subs(T,T_sol), w_C_BC.subs(T,T_sol).evalf())
w_C_CD = N_CD * 4 / EA + N * 4 / EA
display(w_C_CD, w_C_CD.evalf())
display(w_C_CD.subs(N,N_sol), w_C_CD.subs(N,N_sol).evalf())

eq1 = sym.Eq(w_C_BC, w_C_CD)
sol = sym.solve(eq1.subs(T,T_sol).subs(N,N_sol), N_CD)[0]
display(sol.evalf())
display(sol.subs(q,q_sol))
```

```python
display(M_B.subs(N_CD,sol).subs(T,0).subs(N,0).subs(q,q_sol))
display(M_BC_hulp.subs(N_CD,sol).subs(T,0).subs(N,0).subs(q,q_sol))
display(M_BC.subs(N_CD,sol).subs(T,0).subs(N,0).subs(q,q_sol))
display(phi_B.subs(N_CD,sol).subs(T,T_sol).subs(N,N_sol).subs(q,q_sol).evalf())
display(phi_A.subs(N_CD,sol).subs(T,T_sol).subs(N,N_sol).subs(q,q_sol).evalf())
display(w_AB.subs(N_CD,sol).subs(T,T_sol).subs(N,N_sol).subs(q,q_sol).evalf())
display(w_CD.subs(N_CD,sol).subs(T,T_sol).subs(N,N_sol).subs(q,q_sol).evalf())
display(w_C_BC.subs(N_CD,sol).subs(T,T_sol).subs(N,N_sol).subs(q,q_sol).evalf())
display(w_C_CD.subs(N_CD,sol).subs(T,T_sol).subs(N,N_sol).subs(q,q_sol).evalf())
```
